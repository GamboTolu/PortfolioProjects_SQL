```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('/kaggle/input/nashville-housing/NashvilleHousing.csv')
```

```python
df.head()
```

```python
def convert_date(date_str):
    try:
        parsed_date = datetime.strptime(date_str, "%B %d, %Y")
        return parsed_date.strftime("%d-%b-%y")
    except Exception:
        return date_str  

df['FormattedDate'] = df['SaleDate'].apply(convert_date)

# Preview the result
df[['SaleDate', 'FormattedDate']].head()
```

```python
df.to_csv('/kaggle/working/NashvilleHousing_Formatted.csv', index=False)
```
